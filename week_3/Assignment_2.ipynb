### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2.

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv('DATA/drinks.csv')
drinks.head()
```

### Step 4. Which continent drinks the most beer on average?

```python
gdrinks = drinks.groupby('continent')
gdrinks['beer_servings'].mean()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
gdrinks['wine_servings'].agg([np.min,np.max,np.mean,np.sum,np.std])
```

### Step 6. Print the mean alcoohol consumption per continent for every column

```python
gdrinks.mean()
```

### Step 7. Print the median alcoohol consumption per continent for every column

```python
gdrinks.median()
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
gdrinks['spirit_servings'].agg([np.mean, np.min,np.max])
```
